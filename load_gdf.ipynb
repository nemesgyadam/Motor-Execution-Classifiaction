```python
import mne
import numpy as np
import pandas as pd

file ="sample.gdf"
```

```python
data = mne.io.read_raw_gdf(file, preload=True)
```

```python
source_file = 'continous_data/Debug/session_06/data.npy'
source_data = np.load(source_file)
arc_channels = [
    "Time",
    "Trigger",
    "C5",
    "C3",
    "C1",
    "C2",
    "C4",
    "C6",
]
info = mne.create_info(ch_names=arc_channels, sfreq=500, ch_types='eeg')
data = mne.EpochsArray([source_data], info)
```

```python
raw = data.raw
print(raw)
print(raw.info)
```

```python
raw.plot_psd(fmax=50)
```
